```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('C:\\Users\\Prasad Rahatekar\\Desktop\\codebasic\\ml\\linear-reg_single_var\\Exercise\\canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline 
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.plot(df.year, df.per_capita_income_USD, color="blue", marker='*')
```

```python
new_df = df.drop('per_capita_income_USD', axis='columns')
new_df.head()
```

```python
per_capita_income = df.per_capita_income_USD
per_capita_income.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, per_capita_income)
```

```python
reg.predict([[2020]])
```
